```python
df_imp_wonan = df
cols = df_imp_wonan.columns.tolist()
cols = cols[:3] + cols[4:] + [cols[3]]
df_imp_wonan=df_imp_wonan[cols]
df_imp_wonan.head()
```

```python
df_imp_wonan['final_status'] = df_imp_wonan['final_status'].map({'dropout':False, 'graduate':True})
```

```python
df_imp_wonan['final_status'].value_counts()
```

```python
cat_cols=[]
float_cols=[]
bool_cols=[]
target_col='final_status'
features = list(df_imp_wonan)
features.remove(target_col)
for c in features:
    print(c, '\t', df_imp_wonan[c].dtype)
    if df_imp_wonan[c].dtype not in ['float64', 'bool']:
        cat_cols.append(c)
        df_imp_wonan[c] = df_imp_wonan[c].astype('category')
    else: 
        if df_imp_wonan[c].dtype == 'bool':
            bool_cols.append(c)
        else:
            float_cols.append(c)
        
newcols = cat_cols+bool_cols+float_cols+[target_col]
print(newcols)
df_imp_wonan=df_imp_wonan[newcols]
```

```python
for c in list(df_imp_wonan):
    print(c, '\t', df_imp_wonan[c].dtype)
```

```python
cat_cols=[]
num_cols=[]
for c in list(df_imp_wonan)[:-1]:
    if df_imp_wonan[c].dtype.name in ['category', 'bool']:
        cat_cols.append(c)
        print(df_imp_wonan[c].value_counts())
    else:
        num_cols.append(c)
        
class_f = target_col
        
print('Categorical features: ',cat_cols)
print('Numerical features: ',num_cols)
print('Class feature: ',class_f)
        
```

```python
df_imp_wonan_dumm = pd.get_dummies(df_imp_wonan, drop_first=False)
```

```python
for c in list(df_imp_wonan_dumm):
    print(c, '\t', df_imp_wonan_dumm[c].dtype)
```

```python
cat_cols_onehot1=[]
cols_onehot1=list(df_imp_wonan_dumm)
cols_onehot1.remove('final_status')
for c in cols_onehot1:
    if c not in list(df_imp_wonan):
        cat_cols_onehot1.append(c)
        
for c in num_cols:
    if c not in list(df_imp_wonan_dumm):
        print(c)
```

```python
class_f_oh = 'final_status'
        
print('Categorical features: ',cat_cols_onehot1, '\n')
print('Bool features: ',bool_cols, '\n')
print('Numerical features: ',num_cols, '\n')
print('Class feature: ',class_f_oh, '\n')
```

```python
features =  cat_cols_onehot1 + bool_cols + num_cols

coltypes=[]
for c in features:
    if c in cat_cols_onehot1:
        coltypes.append('category')
    if c in bool_cols:
        coltypes.append('bool')
    if c in num_cols:
        coltypes.append('numeric')

pd.DataFrame({'name': features, 'type': coltypes}).to_csv('feature_types_final_simul.csv', sep=';', index=False)
```
